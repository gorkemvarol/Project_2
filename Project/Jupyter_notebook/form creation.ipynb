```python
import pandas as pd
import random
import calendar
import datetime
import csv
```

```python
school_csv_path = "../data/university_lat_lng_output.csv"
subject_csv_path = '../data/majors.csv'

subject_df = pd.read_csv(subject_csv_path)
school_df = pd.read_csv(school_csv_path)
state_names = ["Alaska", "Alabama", "Arkansas", "Arizona", "California", "Colorado", "Connecticut", "District  of Columbia", "Delaware", "Florida", "Georgia", "Hawaii", "Iowa", "Idaho", "Illinois", "Indiana", "Kansas", "Kentucky", "Louisiana", "Massachusetts", "Maryland", "Maine", "Michigan", "Minnesota", "Missouri", "Mississippi", "Montana", "North Carolina", "North Dakota", "Nebraska", "New Hampshire", "New Jersey", "New Mexico", "Nevada", "New York", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Virginia", "Vermont", "Washington", "Wisconsin", "West Virginia", "Wyoming"]
course_levels = [100,200,300,400,500,600,700,800]
group_type = ['Partner', 'Group']
school_df
```

```python
def generate_random_date():
    today = datetime.date.today()

    day    = today.day
    year   = today.year
    month  = today.month
    year  = random.randint(year, year +1)
    month = random.randint(month, 12)

    date_range = calendar.monthrange(year, month)[1]
    day   = random.randint(day, date_range)
    return str(datetime.date(year, month, day))

generate_random_date()
```

```python
school_list = school_df['location_name'].to_list()
school_lat_list = school_df['lat'].to_list()
school_lng_list = school_df['lng'].to_list()
subject_list = subject_df['0'].to_list()
print(len(school_list))
print(len(subject_list))
```

```python
form_input_list = []
for i in range(1000):
    input_dict = {}
    school_selector = random.randint(0,len(school_list)-1)
    subject_selector = random.randint(0,len(subject_list)-1)
    state_selector = random.randint(0,50)
    level_selector = random.randint(0,7)
    group_type_selector = random.randint(0,1)
    input_dict['school'] = school_list[school_selector]
    input_dict['state'] = state_names[state_selector]
    input_dict['subject'] = subject_list[subject_selector]
    input_dict['course_level'] = course_levels[level_selector]
    input_dict['time'] = generate_random_date()
    input_dict['location'] = [school_lat_list[school_selector],school_lng_list[school_selector]]
    input_dict['group_type'] = group_type[group_type_selector]
    form_input_list.append(input_dict)
form_input_list
```

```python
keys = form_input_list[0].keys()
with open('form_inputs.csv', 'w', newline='')  as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(form_input_list)
```
