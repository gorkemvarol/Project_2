```python
# create a params dict that will be updated with new city each iteration
params = {"key": gkey}
#params['key'] = gkey
base_url = "https://maps.googleapis.com/maps/api/geocode/json"

# Loop through the university_pd and run a lat/long search for each adress
for index, row in universities_pd.iterrows():
    #if(index < 5):
    address = row['address']
    # update address key value
    params['address'] = f"{address}"
    print(address)
    
    # make request
    #universities_lat_lng = requests.get(base_url, params=params)
    
    # convert to json
    #universities_lat_lng_json = universities_lat_lng.json()

    try:
        # make request
        universities_lat_lng = requests.get(base_url, params=params)
       
        #check for the response 200 then proceed,
        if universities_lat_lng is not None and universities_lat_lng.status_code == 200:
            # convert to json
            universities_lat_lng_json = universities_lat_lng.json()
            universities_pd.loc[index, "lat"] = universities_lat_lng_json["results"][0]["geometry"]["location"]["lat"]
            universities_pd.loc[index, "lng"] = universities_lat_lng_json["results"][0]["geometry"]["location"]["lng"]
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        print("----------")
        
universities_pd.head()
```
